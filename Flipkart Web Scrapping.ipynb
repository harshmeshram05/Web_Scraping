```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/search?q=mobiles%20under%2050000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
# To check status code
r = requests.get(url)
print(r)
```

```python
soup = BeautifulSoup(r.text, "lxml")
print(soup)
```

```python
# (np -> next page) To get the link of next page
np = soup.find("a",class_="_1LKTO3").get("href")
print(np)
```

```python
# (cnp -> complete next page) 
cnp = "https://www.flipkart.com"+np
print(cnp)
```

```python
for i in range(1,10):
    url = "https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml")
    np = soup.find("a",class_="_1LKTO3").get("href")
    cnp = "https://www.flipkart.com"+np
    print(cnp)
```

```python
box = soup.find("div",class_="_1YokD2 _3Mn1Gg")
```

```python
Product_names = []
names = box.find_all("div",class_="_4rR01T")

for i in names:
    name = i.text
    Product_names.append(name)
print(len(Product_names))
```

```python
Prices = []
price = box.find_all("div",class_="_30jeq3 _1_WHN1")

for i in price:
    name = i.text
    Prices.append(name)
print(Prices)
```

```python
Description = []
desc = box.find_all("ul",class_="_1xgFaf")

for i in desc:
    name = i.text
    Description.append(name)
    
print(Description)
```

```python
Review = []
reviews = box.find_all("div",class_="_3LWZlK")

for i in reviews:
    name = i.text
    Review.append(name)
    
print(Review)
```

```python
for i in range(2,12):
    url = "https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml")
    box = soup.find("div",class_="_1YokD2 _3Mn1Gg")
    Product_names = []
    names = box.find_all("div",class_="_4rR01T")

    for i in names:
        name = i.text
        Product_names.append(name)
    
    Prices = []
    price = box.find_all("div",class_="_30jeq3 _1_WHN1")

    for i in price:
        name = i.text
        Prices.append(name)
    
    Description = []
    desc = box.find_all("ul",class_="_1xgFaf")

    for i in desc:
        name = i.text
        Description.append(name)
    
    Review = []
    reviews = box.find_all("div",class_="_3LWZlK")

    for i in reviews:
        name = i.text
        Review.append(name)
```

```python
df2 = pd.DataFrame({"Product Name":Product_names,"Prices":Prices,"Description":Description,"Review":Review})
print(df2)
```

```python
df2.to_csv("flipkart_mobiles_under-50000.csv")
```
